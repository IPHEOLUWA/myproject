```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/data/world_happiness.csv')
clean = df.dropna()
```

```python
clean.head(10)
```

```python
ssp = clean['Social support'].head(10)
hle = clean['Healthy life expectancy at birth'].head(10)
hle
ssp
```

```python
hle
```

```python
nssp = np.array(ssp)
nhle = np.array(hle)

t_test,p_value = stats.ttest_ind(nssp,nhle,equal_var=True)

print(t_test)
print(p_value)
```

```python
t_test2,p_value2 = stats.ttest_ind(nssp,nhle)
print(t_test2)
print(p_value2)
```

```python
df = 18
alpha = 0.05
t_critical = stats.t.ppf(1-alpha/2,df)
print(t_critical)
```
