```python
df= pd. read_csv('/content/drive/My Drive/Colab Notebooks/data/world_happiness.csv')
clean= df.dropna()
lifeladder= clean['Life Ladder']
pm = np.mean ([lifeladder])
print(pm)
sample= np.array(lifeladder.head(10))
t_test,p_value = stats.ttest_1samp(sample, pm)
print(t_test)
print (p_value)

df=9
alpha=0.05
t_critical = stats.t.ppf(q=1-alpha, df=df)
print(t_critical)
```

```python
if (t_critical<t_test):
  print("Reject the null hypothesis")
else:
  print(" reject the null hypothesis")
```
